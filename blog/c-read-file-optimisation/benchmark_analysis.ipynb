# C `read_file` Optimisation Benchmark

```python
%load_ext lab_black

import csv

import matplotlib.pyplot as plt
import matplotlib.style as style

# configure global plot styles
style.use("seaborn-poster")
style.use("ggplot")

plot_data = {"original": {"x": [], "y": []}, "improved": {"x": [], "y": []}}

with open("2019-09-23.21:49:44-results.csv") as f:
    for row in csv.DictReader(f):
        plot_data[row["read_line_version"]]["x"].append(row["input_file"].split(".")[0])
        plot_data[row["read_line_version"]]["y"].append(float(row["time"]))
```

```python
plt.figure(figsize=(16, 8))

for version in plot_data:
    plt.plot(plot_data[version]["x"], plot_data[version]["y"], label=version)

plt.xlabel("File Size (bytes)")
plt.xticks(rotation="vertical")
plt.ylabel("Time (s)")
plt.legend(loc="upper left")
plt.show()
```
